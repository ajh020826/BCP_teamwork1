```python
from urllib.request import urlopen
url = 'https://movie.daum.net/ranking/reservation'
htmlData = urlopen(url).read()
htmlData = htmlData.decode('utf8')

from bs4 import BeautifulSoup as bs

html = bs(htmlData, 'html.parser')

ol = html.select_one('.infobox')

li=ol.find_all('li')


for item in li:
  rank = int(item.select_one('.rank_num').text)
  mvname = item.select_one('.link_txt').text
  tag = item.select_one('.txt_tag').text.strip()[:3]
  grade=float(item.select_one('.txt_grade').text)
  num=float(item.select_one('.info_txt > .txt_num').text[:-1])
  mvdate = '20' + item.select_one('.txt_info > .txt_num').text.replace('.','-')
  line = f'{rank},{mvname},{tag},{grade}, {num}, {mvdate}\n'
  print(line)
  
```

```python
from urllib.request import urlopen

url = 'https://shopping.interpark.com/best/main.do'

htmlData = urlopen(url).read()

htmlData=htmlData.decode('utf8')

from bs4 import BeautifulSoup as bs

html = bs(htmlData, 'html.parser')

name = html.select('.name')
num = html.select('.number')


for i in range (len(name)):
  price = num[i].text.replace(',','')
  line = f'{name[i].text}, {int(price)}'
  print(line)
```
